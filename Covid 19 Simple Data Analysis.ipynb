```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Importing the dataset.

```python
covid_data = pd.read_csv('Confirmed_Dataset.csv')
```

### We need to remove the unnecessary columns

```python
# Dropping 'Lat' and 'Long' columns
covid_data.drop(['Lat', 'Long'], axis = 1, inplace = True) #axis = 1 is for columns
```

```python
# View of dataset just for surity
covid_data.head(3)
```

### Since, there are many columns having labels as dates (example: 1/22/20 (third column)), we are not sure if they are reduntant or important.

### So, we need to group our data according to a single attribute.

```python
covid_data['Province/State'].unique()
```

```python
covid_data['Country/Region'].unique()
```

#### From the above 2 attributes we see that it would be better if we group our data according to 'Country/Region' as 'Province/State' is bascically a part or consists in a country.

### Grouping the dataset by the 'Country/Region' which will also make it the index.

```python
covid_data_agg = covid_data.groupby('Country/Region').sum()
```

```python
covid_data_agg.head()
```

### Still, we are not sure about the other columns. Let's start with the visualization to make it easy to understand.

### Firstly, we will plot the simple line graph for China.

```python
covid_data_agg.loc['China'].plot(kind = 'line');
```

#### We can also merge the line plots of different countries in the same plot for comparison.

```python
covid_data_agg.loc['China'].plot(kind = 'line');
covid_data_agg.loc['India'].plot(kind = 'line');
covid_data_agg.loc['Italy'].plot(kind = 'line');
covid_data_agg.loc['United Kingdom'].plot(kind = 'line');
covid_data_agg.loc['Australia'].plot(kind = 'line');
covid_data_agg.loc['Germany'].plot(kind = 'line');
plt.legend();
```

```python
covid_data_agg.loc['China'].diff().plot();
covid_data_agg.loc['India'].diff().plot();
covid_data_agg.loc['Italy'].diff().plot();
covid_data_agg.loc['United Kingdom'].diff().plot();
covid_data_agg.loc['Australia'].diff().plot();
covid_data_agg.loc['Germany'].diff().plot();
plt.legend();
```

## Finding the maximum infection rate for all countries and appending it in the aggregated dataset.

```python
countries = list(covid_data_agg.index)
max_rate = []
for i in countries:
    max_rate.append(covid_data_agg.loc[i].diff().max())
```

```python
covid_data_agg['max_infection_rate'] = max_rate
```

```python
covid_data_agg.head()
```

### Changing the index of 'max_infection_rate' column from last to first.

```python
cols = covid_data_agg.columns.tolist()
```

```python
columns = [cols[-1]] + cols[:-1]
```

```python
covid_data_agg = covid_data_agg[columns]
```

```python
covid_data_agg.head()
```

### For the better view.

```python
covid_data_temp = pd.DataFrame(covid_data_agg['max_infection_rate'])
```

```python
covid_data_temp.head()
```

## We want to analyze the impact of Covid-19 on the countries with respect to their GDP or country's citizens' health life expectancy.

### Importing the 'World Happiness Report' file.

```python
happ = pd.read_csv('Worldwide_Happiness.csv')
```

```python
happ.head()
```

```python
happ.info()
```

### There are no null values.

### Getting rid of unnecessary columns.

```python
# Drop columns - social support, perception of corruption
happ.drop(['Score', 'Social support', 'Freedom to make life choices', 'Perceptions of corruption'], axis = 1, inplace=True)
happ.set_index('Country or region', inplace = True)
happ.head()
```

## Performing inner join.

#### An inner join requires each row in the two joined dataframes to have matching column values which is also similar to the intersection of two sets.

```python
final_data = covid_data_temp.join(happ, how = 'inner')
final_data.head()
```

```python
final_data.shape
```

```python
sns.heatmap(final_data.corr());
```

### Visualizing The Results

__1. GDP vs. max_infection_rate__

```python
x = final_data['GDP per capita']
y = final_data['max_infection_rate']
sns.scatterplot(x,y);
```

### We observe that since GDP is given in decimal values and max_infection_rate is in thousands. Hence, we will have to use log scaling to narrow down the values.

```python
x = final_data['GDP per capita']
y = final_data['max_infection_rate']
sns.scatterplot(x, np.log(y)); #apply logscaling to y
sns.regplot(x, np.log(y));
```

### We observe that there is a positive correlation between 'GDP' and 'max_infection_rate' which shows that the people in developed countries are more prone to the Corona virus.

 __2. Healthy life expectancy vs. max_infection_rate__

```python
x = final_data['Healthy life expectancy']
y = final_data['max_infection_rate']
sns.scatterplot(x, np.log(y)); #apply logscaling to y
sns.regplot(x, np.log(y));
```

### We observe that there is a weak positive correlation between 'Healthy life expectancy' and 'max_infection_rate' which shows that as life expaectancy increases, infection rate also increases.
